```python
import numpy as np
```

```python
def get_tmux_bash_script(i_VM, i_threads, i_FI_loop_ranges, i_variant, i_vinputs, i_bit_rank_strategy, i_num_bits_to_protect):
    bash_string  = "#!/bin/bash\n"

    for i in range(i_VM*i_threads, i_VM*i_threads+i_threads):
        # bash_string += f"tmux new-session -d -s vs{i} './scripts/sampling-faulty-eval-experiment-forloop.sh {i}'\n"

        lbi = i_FI_loop_ranges[i][0]
        hbi = i_FI_loop_ranges[i][1]
        model_name = f'"ECONT-{i_variant}"'
        bash_string += f"tmux new-session -d -s vs{i:03} './scripts/iccad_2023_experiment3_FI_forloop_{i_variant}.sh {model_name} {i_vinputs} {i:03} {lbi} {hbi} 64 {i_bit_rank_strategy} {i_num_bits_to_protect}'\n"

    return bash_string

def gen_EXP3_ECONT_VM_scripts(i_bits, i_variant, i_vinputs, i_bit_rank_strategy, i_num_bits_to_protect):
    total_bits = i_bits
    vsystems_per_VM = 16
    total_VMs = 13
    total_vsystems = vsystems_per_VM * total_VMs

    bits_per_vsystem = np.ceil(total_bits/total_vsystems)

    ranges_for_VM_scripts = list()
    for curr_vs in range(total_vsystems):
        curr_lbi = int(bits_per_vsystem*curr_vs)
        curr_hbi = int(min(curr_lbi + bits_per_vsystem, total_bits))  
        ranges_for_VM_scripts.append( (curr_lbi, curr_hbi, curr_hbi-curr_lbi))


    for curr_vm in range(total_VMs):
        with open(f"iccad_2023_experiment3_{i_variant}_{i_bit_rank_strategy}_{i_num_bits_to_protect}_VM{curr_vm:02}.sh", "w") as f:
            f.write(get_tmux_bash_script(curr_vm, vsystems_per_VM, ranges_for_VM_scripts, i_variant, i_vinputs, i_bit_rank_strategy, i_num_bits_to_protect))


def gen_EXP3_ECONT_VM_scripts_range(i_bits, i_variant, i_vinputs, i_bit_rank_strategy, i_num_bits_to_protect):
    total_bits = i_bits
    vsystems_per_VM = 16
    total_VMs = 13
    total_vsystems = vsystems_per_VM * total_VMs

    bits_per_vsystem = np.ceil(total_bits/total_vsystems)

    ranges_for_VM_scripts = list()
    for curr_vs in range(total_vsystems):
        curr_lbi = int(bits_per_vsystem*curr_vs)
        curr_hbi = int(min(curr_lbi + bits_per_vsystem, total_bits))  
        ranges_for_VM_scripts.append( (curr_lbi, curr_hbi, curr_hbi-curr_lbi))

    return ranges_for_VM_scripts

#I: Total bits per ECON-T variant
### 61344 large pareto
### 12720 baseline
### 10240 small pareto
for bits_to_protect in [0, 8080]:
    gen_EXP3_ECONT_VM_scripts(14000, "baseline", "v20000-h8192", "oracle", bits_to_protect)
    gen_EXP3_ECONT_VM_scripts(14000, "baseline", "v20000-h8192", "hessian", bits_to_protect)
```
